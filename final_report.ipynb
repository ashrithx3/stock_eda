```python
import matplotlib.pyplot as plt
import numpy as np

# Assets and R² values
assets = ["UBER", "ETH", "AAPL", "NFLX"]
poly_r2 = [0.4327, 0.4327, 0.2165, 0.2092]

# More natural, human-friendly colors
colors = ["#0ae4f3",  # muted blue
          "#ff0eb7",  # muted orange
          "#2ca02c",  # muted green
          "#faf202"]  # muted red

plt.figure(figsize=(10, 6))
bars = plt.bar(assets, poly_r2, color=colors)

plt.title("Polynomial Model R² Comparison Across Assets", fontsize=16, fontweight="bold")
plt.xlabel("Asset", fontsize=13)
plt.ylabel("R² Value", fontsize=13)
plt.ylim(0, 0.5)
plt.grid(axis='y', alpha=0.3)

# Add values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01,
             f"{yval:.3f}", ha='center', fontsize=12)

plt.show()
```

**Notes:**

---

## Best and Worst Fits

- **Best fit overall:** ETH (R² ≈ 0.43) → strongest relationship between volume and volatility.  
- **Worst fit overall:** NFLX (R² ≈ 0.21) → weakest relationship, high noise.

**Stocks:** UBER, AAPL, and NFLX show low to moderate R² (~0.21–0.43).  
**Crypto:** ETH shows highest R² (~0.43); polynomial only slightly improves fit.

---

## Final Discussion

The analysis indicates that **volume changes (ΔV) influence volatility changes (Δσ) but vary across assets**:

- **Stocks:** Mostly linear, low R² → volume is not dominant. Stock-specific factors (liquidity, news, behavior) matter more.  
- **Crypto:** ETH shows highest sensitivity to volume (R² ≈ 0.43). Lower liquidity and speculative activity increase the effect.  

Polynomial models did not significantly outperform linear models, except for slight gains in UBER and ETH. **Linear models are sufficient for interpretation**, but additional factors—market sentiment, liquidity, macro conditions—also play crucial roles.

```python
import matplotlib.pyplot as plt
import numpy as np

# Assets and R² values
assets = ["UBER", "ETH", "AAPL", "NFLX"]
linear_r2 = [0.428, 0.4279, 0.2165, 0.2083]
poly_r2 = [0.4327, 0.4327, 0.2165, 0.2092]
best_fit = ["Poly", "Poly", "Same", "Same"]

# Human-friendly colors
colors = ["#0ae4f3", "#ff0eb7", "#2ca02c", "#faf202"]

plt.figure(figsize=(10, 6))
bars = plt.bar(assets, poly_r2, color=colors, alpha=0.8)

plt.title("Polynomial Model R² Comparison Across Assets", fontsize=16, fontweight="bold")
plt.xlabel("Asset", fontsize=13)
plt.ylabel("R² Value", fontsize=13)
plt.ylim(0, 0.5)
plt.grid(axis='y', alpha=0.3)

# Add values and Best Fit annotations
for i, bar in enumerate(bars):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01,
             f"{yval:.3f}", ha='center', fontsize=12)
    plt.text(bar.get_x() + bar.get_width()/2, yval - 0.05,
             f"{best_fit[i]}", ha='center', fontsize=11, color='black')

plt.show()
```
